```python
images = []
def read_images(data):
    for i in range(len(data)):
        img = cv2.imread(data[i])
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        img = cv2.resize(img,(224,224))
        images.append(img)
    return images
```

### (2) Image enhancement


### Method 1

```python
def dft(img,k):
    img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    f = np.fft.fft2(img)
    fshift = f*pow(np.abs(f),k) 
    img_back = np.fft.ifft2(fshift)
    img_back = np.real(img_back)
    img_back = img_back.astype(np.uint8)
    plt.imshow(img_back,cmap = 'gray')
    return img_back
```

```python
img_dft = dft(images[0],0)
```

```python
def histEqualization(img):
    im = img.copy()
    size = im.shape[0]*im.shape[1]
    cdf = np.zeros(256)
    for i in range(0,256):
        cdf[i]=np.round((np.count_nonzero(im<=i)/size)*255)
    
    im = cdf[im]
    im = im.astype(np.uint8)
    return im
```

```python
img_final = histEqualization(img_dft)
```

```python
plt.imshow(img_final,cmap='gray')
```

### (3) Binarization

### (4) Thinning 


```python
def erosion(image, size):
    img = image.copy()
    kernel = np.ones((size,size), dtype=np.uint8)
    img_op = np.zeros(img.shape)
    n = size // 2
      
    for r in range(n,img.shape[0]-n):
      for c in range(n,img.shape[1]-n):
            img_op[r][c] = np.min(img[r-n:r+n+1,c-n:c+n+1]*kernel)
  
    return img_op.astype(np.uint8)
```

```python
def dilation(image, size):
    img = image.copy()
    kernel = np.ones((size,size), dtype=np.uint8)
    img_op = np.zeros(img.shape)
    n = size // 2
      
    for r in range(n,img.shape[0]-n):
      for c in range(n,img.shape[1]-n):
            img_op[r][c] = np.max(img[r-n:r+n+1,c-n:c+n+1]*kernel)
  
    return img_op.astype(np.uint8)
```

```python
img1 = dilation(img_final,3)
img2 = erosion(img1,3)
img3 = dilation(img2,3)
```

```python
plt.imshow(img3,cmap='gray')
```

### (5) Feature Extraction

```python
def feature_extraction(img):
    im = img.copy()
    for r in range(1,img.shape[0]-1):
      for c in range(1,img.shape[1]-1):
            sum = np.sum(img[r-1:r+2,c-1:c+2])
            if sum ==5 or sum == 7 or sum == 4:
                im[r][c]= 0
            else:
                im[r][c]=255
    return im
                
```

```python
features = feature_extraction(img3)
```

```python
plt.imshow(features,cmap='gray')
```
